```python
import requests

def get_categories():
    url = "https://www.getonbrd.com/api/v0/categories?per_page=10&page=1"
    response = requests.get(url)
    if response.status_code == 200:
        jobs = response.json()
        return jobs
    else:
        return f"Error: {response.status_code}"


jobs_data = get_categories()
jobs_data
```

```python
import json

def get_all_jobs():
    url = "https://www.getonbrd.com/api/v0/categories/programming/jobs?per_page=100&page=1"
    response = requests.get(url)
    if response.status_code == 200:
        jobs = response.json()
        return jobs
    else:
        return f"Error: {response.status_code}"

def save_jobs_to_json(jobs_data, file_name):
    with open(file_name, 'w', encoding='utf-8') as file:
        json.dump(jobs_data, file, ensure_ascii=False, indent=4)

if __name__ == "__main__":
    jobs_data = get_all_jobs()
    if isinstance(jobs_data, dict):  # Only save if the data is a dictionary (JSON)
        save_jobs_to_json(jobs_data, '../data/jobs_data.json')
```
